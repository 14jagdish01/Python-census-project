```python
print("Welcome my Game Tic tac toi Im Jack")
def print_board(board):
    print("   |   |")
    print(" " + board[0] + " | " + board[1] + " | " + board[2])
    print("   |   |")
    print("-----------")
    print("   |   |")
    print(" " + board[3] + " | " + board[4] + " | " + board[5])
    print("   |   |")
    print("-----------")
    print("   |   |")
    print(" " + board[6] + " | " + board[7] + " | " + board[8])
    print("   |   |")

def check_win(board):
    win_combinations = [(0, 1, 2), (3, 4, 5), (6, 7, 8),
                        (0, 3, 6), (1, 4, 7), (2, 5, 8),
                        (0, 4, 8), (2, 4, 6)]
    for combination in win_combinations:
        if board[combination[0]] == board[combination[1]] == board[combination[2]] != " ":
            return board[combination[0]]
    if " " not in board:
        return "Tie"
    return None

def get_move(player, board):
    while True:
        move = input(f"{player}'s turn. Enter a position (1-9): ")
        if not move.isdigit():
            print("Invalid input. Please enter a number between 1 and 9.")
        elif int(move) < 1 or int(move) > 9:
            print("Invalid input. Please enter a number between 1 and 9.")
        elif board[int(move) - 1] != " ":
            print("That position is already occupied. Please choose another one.")
        else:
            return int(move) - 1

def play_game():
    board = [" "] * 9
    current_player = "X"
    winner = None
    while not winner:
        print_board(board)
        move = get_move(current_player, board)
        board[move] = current_player
        winner = check_win(board)
        if current_player == "X":
            current_player = "O"
        else:
            current_player = "X"
    print_board(board)
    if winner == "Tie":
        print("It's a tie!")
    else:
        print(f"{winner} wins!")

play_game()
```
